# Web Scrapping

```python
from bs4 import BeautifulSoup
import requests
import pandas
from pandas import Series, DataFrame

url="https://www.flipkart.com/?semcmpid=sem_F1167BY7_Brand_adcenter&s_kwcid=AL!739!10!76347463873717!76347484024120&ef_id=Xwhz-AAABf4LtiiE:20200710135820:s"

```

```python
result=requests.get(url)
print(result.content)
```

```python
soup=BeautifulSoup(result.content,'lxml')
print(soup)
```

```python
content=soup.findAll('div')
print(content)
```

```python
x=content[1].find(class_="_1T-JyI")
print(x)
l=x.findAll('img')
print(l)
```

```python
series=[ele.get_text() for ele in l]
print(series)
```
