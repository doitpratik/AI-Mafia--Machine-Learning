```python
x_test = dfxtest.values
y_train = dfytrain.values
x_train = dfxtrain.values
```

```python
# KNN Algo
def dist(x1, x2):
    return np.sqrt(sum((x1-x2)**2))

def knn(X,Y,querypoint,k=5):
    
    vals=[]
    m = X.shape[0]
    
    for i in range(m):
        #compute distance
        d = dist(querypoint, X[i])
        vals.append((d, Y[i]))
    #sorting the values in vals list    
    vals = sorted(vals)
    #only taking those values which are given input for nearest value
    vals = vals[:k]
    
    #majority vote
    vals = np.array(vals)
    new_value = np.unique(vals[:,1], return_counts=True)
    
    print(new_value)
    
    index = new_value[1].argmax()
    pred = new_value[0][index]
    return pred
```

```python
#Testing model
pred=[]
for i in range(x_test.shape[0]):
    my_pred=knn(x_train,y_train,x_test[i])
    pred.append(my_pred)
my_df=pd.DataFrame(pred)
```
